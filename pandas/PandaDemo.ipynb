```python
#crating the series 
a=[1,2,3,4,5,6]

#without indexing
s1=pd.Series(a)
print("series: ",s1)

print("\n")

#with indexing
s1=pd.Series(a,index=["i1","i2","i3","i4","i5","i6"]) #length of index must be equal to the length of the array 
print("series with index: ",s1)

print("\n")

#printing individial element
print("i3: ",s1["i3"])
```

```python
# DataFrame will convert the data into the row & column format ie. tabular format
student = {
            "subject": ["AIML","CNN","WS","CS"],
            "marks" : [90,89,70,99]
          }
sStud = pd.DataFrame(student)
print(sStud)

# to fatch single row from DataFrame
print(f"\n{sStud.loc[3]}")
```

```python
# reading data from file
studentData = pd.read_csv('students.csv')
print(f"\n{studentData}")

#printing the first 5 records 
print(f"\n{studentData.head()}") # will give the firts 5 records

#printing the last 5 records 
print(f"\n{studentData.tail()}") # will give the last 5 records

#fatching any single row from data
print(f"\n{studentData.loc[2]}")

#info about the data
print(f"\n{studentData.info()}")

# droping the null data
newStudentData = studentData.dropna()
print(f"\nDrop Null Values:\n{newStudentData}")

#replace value to null values
filledStudentData=studentData.fillna(30)
print(f"\n\nreplace Null Values:\n{filledStudentData}")


#update spacific column
new = studentData["Course"].fillna("MCA")
print(f"\n\nupdating spacific value\n{new}")

studentData.fillna(30,inplace=True) # will rewrite the object
print(f"\n\nRewritted Old Data\n{studentData}")


#update spacific column
new = studentData["Percentage"].fillna(50)
print(f"\n\nupdating spacific value\n{new}")

#finding the mean
mean = studentData["Percentage"].mean()
print(f"\nmean: {mean}")

#median
median = studentData["Percentage"].median()
print(f"\nmedian: {median}")

#mode
mode = studentData["Percentage"].mode()
print(f"\nmode: {mode[0]}")
```
