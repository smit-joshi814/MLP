```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading csv 
movies = pd.read_csv("movies.csv")
movies
```

```python
# scatter diagrams
plt.scatter(movies['vote_average'],movies['total'],s=40) # where s is size
plt.show()
```

```python
# plt.scatter

plt.scatter(movies['vote_average'],movies['total'],c='r',s=40) # where c is color
plt.show()
```

```python
plt.scatter(movies['vote_average'],movies['total'],c='r',s=40,alpha=0.5) # where alpha is opacity
plt.show()
```

```python
plt.scatter(movies['vote_average'],movies['total'],c='r',s=40,alpha=0.5,marker='*') # where marker is use to mark the elements
plt.show()
```

```python
# bar chart
plt.bar(movies['vote_average'],movies['total'])
plt.show()
```

```python
#horizontal bar chart
plt.barh(movies['vote_average'],movies['total'])
plt.show()
```

```python
# bar chart color & width
plt.bar(movies['vote_average'],movies['total'],color='g',width=5)
plt.show()
```

```python
# barh chart color & height
plt.barh(movies['vote_average'],movies['total'],color='g',height=5)
plt.show()
```

```python
# histogram
plt.hist(movies['total'],color='r')
plt.show()
```

```python
#pie chart
# label is used to show the labels
# explode used to extract pie from the chart and show it differently
# shadow is used to show the shadow
# colors is used to shwo the colors for each pie
track_stud = np.array([20,40,30,50,10])
ex=[0,0,0.2,0,0.2]
track_name=np.array(["Web","Data","Mobile","CyberSecurity","IoT"])
colors=["r","g","b","m","pink"]
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=colors)
plt.legend(title="Track Selection",loc="upper left")
plt.show()
```
