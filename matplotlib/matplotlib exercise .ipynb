```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movies.csv')

movies
```

```python
font = {'size':15,'color':'g'}



# plt.grid()
# plt.grid(axis='x')
# plt.grid(axis='y')

plt.subplot(1,2,1)
plt.title('data 1',fontdict=font,loc='left')
plt.plot(movies['total'],movies['average'],'o',ms=10,mec='g',mfc='#ffff',linestyle='dashdot',color='m',linewidth=4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(1,2,2)
plt.title('data 2',fontdict=font,loc='left')
plt.plot(movies['vote_count']/2,movies['total'],marker='X',ms=10,mec='g',mfc='#ffff',linestyle='dotted',color='r',linewidth=2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.suptitle('movies data')
```
