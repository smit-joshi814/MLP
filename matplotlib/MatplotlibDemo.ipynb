```python
import pandas as pd
import numpy as np
# you can use it to create tables
import matplotlib.pyplot as plt
```

```python
# arrays
arr1 = np.array([5,8,10,9,6,7,5,3,2,2,42,4])
arr2 = np.array([5,8,10,9,6,54,5,22,2,2,42,4])

# plotting arrays
plt.plot(arr1,arr2)
```

```python
arr3 = np.array([10,20,30,40,50])
arr4 = np.array([3020,40,20,10,30])

#plt.plot(arr1,arr2)
#plt.plot(arr3,arr4)

plt.plot(arr1,arr2,arr3,arr4)  # to plot all the arrays
```

```python
# plots with dots
plt.plot(arr1,arr2,'o')
```

```python
# marker x
plt.plot(arr1,arr2,marker='X')
```

```python
# marker size ms
plt.plot(arr1,arr2,marker='o',ms=10)
```

```python
# marker color mec
plt.plot(arr1,arr2,marker='x',ms=10,mec='r')
```

```python
# marker front color mfc
plt.plot(arr1,arr2,marker='X',ms=20,mec='r',mfc='b')
```

```python
# linestyle
plt.plot(arr1,arr2,linestyle='dashed')
```

```python
# using multiple linestyle
plt.plot(arr1,arr2,linestyle='dashdot')
```

```python
# color
plt.plot(arr1,arr2,linestyle='dashdot',color='g')
```

```python
# linewidth
plt.plot(arr1,arr2,linestyle='dashdot',color='g',linewidth=10)
```

```python
# font
font={'size':15,'color':'g'}

# title
plt.title('Demo Chart',fontdict=font,loc='right')

#labling
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(arr1,arr2,linestyle=':',color='g',linewidth=4)

# grids
# plt.grid()
plt.grid(axis='x')
# plt.grid(axis='y')
```

```python
#subplotting

plt.subplot(1,2,1) # row column index
plt.title('p1')
plt.plot(arr1,arr2)

plt.subplot(1,2,2)
plt.plot(arr3,arr4)
plt.title('p2')

plt.suptitle('SuperTitle')
```
